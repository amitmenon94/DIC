```r
#problem 6
library("ggmap")
library("maptools")
library(maps)
register_google(key = "your key here")
#added regions
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima","Peru", "Johannesbury", "SA","Mumbai","Buffalo")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col=rainbow(10), bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
#pch changes the point type
points(visit.x,visit.y, col="black", pch=70)
```
